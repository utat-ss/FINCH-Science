```python
import pandas as pd
import numpy as np
import spectral_library # * all major functions have been moved here
import matplotlib.pyplot as plt
```

```python
df = spectral_library.open_file('unmixing/original_data.csv')
fractions, nr900to1700, _ = spectral_library.take_subset(df,start=900,end=1700)
samples = df.shape[0]
```

```python
df.shape[1]
```

```python
npv = fractions['npv_fraction'].to_list()
gv = fractions['gv_fraction'].to_list()
soil = fractions['soil_fraction'].to_list()
```

```python
fig = plt.figure(figsize=(5, 5))
ax = fig.add_subplot(projection='3d')

ax.scatter(npv, gv, soil, marker='o')

ax.set_xlabel('npv')
ax.set_ylabel('gv')
ax.set_zlabel('soil')

plt.show()
```

```python
def simple_histogram(data=[1,2,3],title="Title",x="x-axis",y='y-axis',bins=10):
    fig, ax = plt.subplots(figsize=(5, 3))
    ax.set_xlabel(x)
    ax.set_ylabel(y)
    ax.set_title(title)
    ax.hist(data,bins=bins)
    print()
```

```python
simple_histogram(npv,'npv','fraction','samples',bins=50)
```

```python
simple_histogram(gv,'gv','fraction','samples',bins=50)
```

```python
simple_histogram(soil,'soil','fraction','samples',bins=50)
```
